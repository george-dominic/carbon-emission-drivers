```python
import pandas as pd
```

```python
def fix_data(sheet_name):
  data = pd.read_excel("/content/green data.xlsx", sheet_name=sheet_name)
  df = data.T

  first_row_values = df.iloc[0,:].values

  df_new = df[1:]
  df_new.index = pd.to_datetime(df_new.index).strftime('%Y')
  gdf = df_new.groupby(df_new.index).sum()
  gdf = gdf.loc[['2017','2018','2019','2020','2021']]
  first_row_df = pd.DataFrame([first_row_values], columns=df_new.columns)
  concat_df = pd.concat([first_row_df, gdf])
  concat_df.columns = concat_df.iloc[0,:]
  concat_df.drop(0, inplace = True)
  concat_df.rename_axis(sheet_name, inplace = True)
  return concat_df
```

```python
ebit_df = fix_data("EBIT")
mc_df = fix_data("MC")
oe_df = fix_data("OE")
eps_df = fix_data("EPS")
capex_df = fix_data("CAPEX")
bv_df = fix_data("BV")
roa_df = fix_data("ROA")

roa_df
```

```python

merged_df = pd.DataFrame()

# Iterate over each ticker
for ticker in roa_df.columns:
    # Create a dictionary with the extracted data for the current ticker
    data_dict = {
        'EBIT Data': ebit_df[ticker].to_dict(),
        'OE Data': oe_df[ticker].to_dict(),
        'EPS Data': eps_df[ticker].to_dict(),
        'Capex Data': capex_df[ticker].to_dict(),
        'ROA Data': roa_df[ticker].to_dict()
    }

    # Create a new dataframe from the dictionary
    new_df = pd.DataFrame(data_dict)

    # Add a new level to the column index with the ticker symbol
    new_df.columns = pd.MultiIndex.from_product([[ticker], new_df.columns], names=['Ticker', ''])

    # Concatenate the new dataframe with the merged dataframe
    merged_df = pd.concat([merged_df, new_df], axis=1)

# Sort the column index in alphabetical order of ticker symbols
merged_df = merged_df.sort_index(axis=1)
```

```python
merged_df[0:5].to_excel('gds_multiindex.xlsx')
```
