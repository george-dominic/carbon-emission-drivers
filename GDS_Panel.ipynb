### Time-invariant panel

Transforming time-invariant variables into a panel format

```python
ti_data = pd.read_excel('/content/gds_multiindex.xlsx',sheet_name='Ticker-sector')
```

```python
# =============================================================================
# time invariant variables: Name, Sector, Tickers etc
# =============================================================================
# Get Ticker and Sector columns
ti=ti_data[['Ticker','Sector']].copy()
# Set Ticker as the index
ti.index=ti['Ticker']
ti.drop('Ticker',axis=1,inplace=True)
ti
```

```python
# Transpose the df
ti=ti.T
ti
```

```python
# Duplicate the df
ti_re=pd.concat([ti]*4).reset_index(drop=True)
ti_re
```

```python
from datetime import datetime
import numpy as np
# format sample period: start and end dates
# function: string to datetime
start_date = datetime.strptime("2018-01-01", "%Y-%m-%d")
end_date = datetime.strptime("2022-01-01", "%Y-%m-%d")
```

```python
# generate a yearly dates series
# format it as df
date_list = pd.date_range(start_date, end_date, freq='Y')
dates=pd.DataFrame(date_list,columns=['Dates'])
dates.reset_index(drop=True, inplace=True)
dates
```

```python
# concat two dfs on the common column
ti_re=pd.concat([dates,ti_re],axis=1)
ti_re=ti_re.set_index('Dates')
ti_re
```

```python
# make it into panel format
# name the sector col as Sector
ti_panel=pd.DataFrame(ti_re.unstack())
ti_panel.columns=['Sector']
ti_panel
```

```python
# format the panel & MultiIndex:
# rename the MultiIndex
ti_panel.rename_axis(index=['id','Dates'],inplace=True)
# sort values by MultiIndex
ti_panel=ti_panel.sort_values(['id', 'Dates'], ascending=[True, True])
ti_panel
```

```python
# check the # of firms
ti_panel.index.get_level_values(0).nunique()
```

```python
# check the # of dates
ti_panel.index.get_level_values(1).nunique()
```

```python
# save the formatted panel into parquet
ti_panel.to_parquet('ti_panel.parquet')
```

Financial Data Panel

```python
# =============================================================================
# # FINANCIAL DATA PANEL
# =============================================================================
data=pd.read_excel('/content/gds_multiindex.xlsx',sheet_name='Financials')
data.head()
```

```python
# daeling with 'Unnamed' columns
# forward fill, if starts with 'Unnamed', fill it with nan
data.columns = data.columns.to_series().mask(lambda x: x.str.startswith('Unnamed')).ffill()
data.head()
```

```python
# name the first item in the column header as 'Dates'
data.columns.values[0] = 'Dates'
data.head()
```

```python
# get MultiIndex columns:
# make a copy of the original df
df=data.copy()
df.head()
```

```python
# drop 'Dates' column
df.drop(['Dates'],axis=1,inplace=True)
df.head()
```

```python
# get MultiIndex from arrays of df columns [Tickers] and the name of data fields in the first row [EBIT, BS_TOT_ASSET ...]
mul_cols=pd.MultiIndex.from_arrays([df.columns, df.iloc[0].values])
mul_cols
```

```python
# get values only df
data.head()
```

```python
# drop the first row and reset index
data=data.drop([0]).reset_index(drop=True)
data
```

```python
# set 'Dates' to datetime format
data['Dates']=pd.to_datetime(data['Dates'])
data
```

```python
# set 'Dates' as df index
data.set_index(data['Dates'], inplace=True)
data.drop(['Dates'],axis=1,inplace=True)
data
```

```python
# add previously created mul_cols as data df's columns
data.columns=mul_cols
data.head()
```

```python
# before convert the data df (with multi-level column headers) into a panel
# shift the dates index! becuase we want to use lagged financial data to explain the DV
# check the current datetime index
data.index
```

```python
# shifting the original datetime index one-year forward
data.index = data.index + pd.tseries.offsets.YearEnd()
data.index
```

```python
# slice the df based on datetime index
data=data[:'2021-12-31']
data
```

```python
# convert the data df (with Multi-level columns) into panel
fin_panel=data.stack(0,dropna=False)
fin_panel.head()
```

```python
# rename the MultiIndex
fin_panel.rename_axis(index=['Dates', 'id'],inplace=True)
# swap the levels of index, since we want to have id-dates like index
fin_panel=fin_panel.swaplevel()
# sort values by MultiIndex
fin_panel=fin_panel.sort_values(['id', 'Dates'], ascending=[True, True])
# set the df value format as float
fin_panel=fin_panel.astype(float)
fin_panel.head()
```

```python
# save formatted financial data panel into parquet
fin_panel.to_parquet('fin_panel.parquet')
```

### Emmissions data Panel

```python
# =============================================================================
# # EMISSION DATA PANEL
# =============================================================================
# emission data to panel format
emission=pd.read_excel('/content/gds_multiindex.xlsx',sheet_name='Emmissions')
emission.head()
```

```python
#  forward fill, if starts with 'Unnamed', fill it with nan
emission.columns = emission.columns.to_series().mask(lambda x: x.str.startswith('Unnamed')).ffill()
emission.head()
```

```python
# get multi-columns
# get a copy of original emissions df
df=emission.copy()
df.head()
```

```python
# set the NaN value in the header as Dates
df.columns.values[0] = 'Dates'
df.head()
```

```python
# drop the 'Dates' column
df.drop(['Dates'],axis=1,inplace=True)
# set up two levels of multi columns: df's column - headers [Dates, Ticker 1, Ticker 2...] & the first row [Dates, Scoep 1...]
mul_cols=pd.MultiIndex.from_arrays([df.columns, df.iloc[0].values])
mul_cols
```

```python
# get emission values only df
emission.head()
```

```python
# drop the first row and reset the index
emission=emission.drop([0]).reset_index(drop=True)
emission.head()
```

```python
# format the 'Dates' as datetime
emission['Dates']=pd.to_datetime(emission['Dates'])
```

```python
# and set 'Dates' column as index
emission.set_index(emission['Dates'],inplace=True)
emission.head()
```

```python
# drop the 'Dates' column
emission.drop(['Dates'],axis=1,inplace=True)
```

```python
emission.head()
```

```python
# set the previously created multi-level columns (i.e., headers) as columns
emission.columns=mul_cols
emission.head()
```

```python
# convert the emission df that with two levels of columns (i.e., headers) into panel format
emission_panel=emission.stack(0,dropna=False)
emission_panel.head()
```

```python
# rename the MultiIndex
emission_panel.rename_axis(index=['Dates', 'id'],inplace=True)
# swap the index levels
emission_panel=emission_panel.swaplevel()
# sort values by MultiIndex
emission_panel=emission_panel.sort_values(['id', 'Dates'], ascending=[True, True])
#
emission_panel.head()
```

```python
# set the values as float
emission_panel=emission_panel.astype(float)
```

```python
# save the formatted emission panel to parquet
emission_panel.to_parquet('emission_panel.parquet')
```

```python
import statsmodels.api as sm
from linearmodels.panel import PanelOLS
```

```python
# create Y df
Y = panel[['Scope_1']]
Y
```

```python
# specify the model
mod = PanelOLS(Y, X)
res = mod.fit()
print(res)
```

It's worth noting that the p-values for const, Capex Data, EBIT Data, EPS Data, and OE Data are relatively high, indicating that these variables may not be statistically significant in explaining the variation in the dependent variable. However, the p-value for ROA Data is very low (p < 0.001), suggesting that it is a statistically significant variable.

The F-statistic is 6.9433, with a corresponding p-value of 0.0000, indicating that at least one of the independent variables in the model is jointly significant.

Overall, this summary provides an overview of the regression results and the statistical significance of the variables in explaining the variation in the dependent variable.

This means that, holding other variables constant, an increase in these independent variables is associated with a decrease in the dependent variable (Scope_1).

Specifically:

An increase in Capex Data by one unit is associated with a decrease in Scope_1 by -5.356e-06 units.
An increase in EBIT Data by one unit is associated with a decrease in Scope_1 by -1.093e-05 units.
An increase in EPS Data by one unit is associated with a decrease in Scope_1 by -436.55 units.
An increase in ROA Data by one unit is associated with a decrease in Scope_1 by -1.014e+06 units.
However, the parameter estimate for OE Data is positive, indicating that an increase in OE Data by one unit is associated with an increase in Scope_1 by 7.203e-07 units.
